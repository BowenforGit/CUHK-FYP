```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
sns.set_context("paper")
```

```python
work=["ART_INS","ART_BULK","ART_LOOKUP","ART_RANGE","ART_WITHOUT",\
      "GRASPER_INS","GRASPER_BULK","GRASPER_LOOKUP","GRASPER_RANGE","GRASPER_WITHOUT"]

data_type={
    "0": "Type-I",
    "1": "Type-II",
    "2": "Type-III"
}

size=["1000","10000","100000","1000000"]

rep = 5
```

```python
# Experiment 1
# ART insertion vs. ART bulk loading vs. Grasper insertion

methods = ["ART_INS","ART_BULK","GRASPER_INS"]

for input_type in ["0","1","2"]:
    df_exp1 = pd.DataFrame(columns=["Methods","Data Size","Throughput"])
    for task in methods:
        for s in size:
            fname="./results/"+task+"/"+task+"_"+s+"_"+input_type+".txt"
            avg = 0
            with open(fname,"r") as f:
                for line in f.readlines():
                    avg+=float(line)
            avg /= rep
            res = {
                "Methods": task,
                "Data Size" : s,
                "Throughput" : avg
            }
            df_exp1 = df_exp1.append(res,ignore_index=True)
    ax = plt.figure()
    sns.barplot(x="Data Size", y="Throughput", hue="Methods", data=df_exp1)
    plt.savefig("Insertion_"+input_type+".png",dpi=400)
```

```python
# Experiment 3
# ART RANGE vs. GRASPER RANGE

methods = ["ART_RANGE","GRASPER_RANGE"]

for input_type in ["0","1","2"]:
    df_exp1 = pd.DataFrame(columns=["Methods","Data Size","Time(ms)"])
    for task in methods:
        for s in size:
            fname="./results/"+task+"/"+task+"_"+s+"_"+input_type+".txt"
            avg = 0
            with open(fname,"r") as f:
                for line in f.readlines():
                    avg+=float(line)*1000
            avg /= rep
            res = {
                "Methods": task,
                "Data Size" : s,
                "Time(ms)" : avg
            }
            df_exp1 = df_exp1.append(res,ignore_index=True)
    ax = plt.figure()
    sns.barplot(x="Data Size", y="Time(ms)", hue="Methods", data=df_exp1)
    plt.savefig("RANGE_"+input_type+".png",dpi=400)
```
